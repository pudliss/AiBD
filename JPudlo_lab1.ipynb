Ćwiczenie 3 wprowadzające - funckja do plotowania wykresów

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_function(first,last):
    x= np.arange(first, last, 0.01)
    y=x*x+5
    plt.plot(x,y,label="y=x^2+5")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.suptitle("x>"+str(first)+",x<"+str(last))
    plt.legend()
    plt.show()
    
```

```python
plot_function(-1,1)
plot_function(-6,6)
plot_function(0,5)
```
